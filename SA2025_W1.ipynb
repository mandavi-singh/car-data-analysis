### 1. What is name of car that has the highest horsepower?  

```python
df.reset_index(inplace=True)

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Find the row with max horsepower
max_hp_row = df[df['horsepower'] == df['horsepower'].max()]

# Display the name
print(max_hp_row['name'].values[0])
```

### 2. How many cars have mpg ≥ 35?

```python
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

high_mpg_count = df[df['mpg'] >= 35].shape[0]
print(high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# Filter the dataframe
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_df['origin'].mode()[0]
print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
df['acceleration'] = pd.to_numeric(df['acceleration'], errors='coerce')

# Filter rows where origin is Japan
japan_cars = df[df['origin'] == 'japan']

# Calculate mean acceleration
mean_acc = japan_cars['acceleration'].mean().round(2)
print(mean_acc)
```

### 5. Which year had the highest average mpg?

```python
# Group by model_year and calculate average mpg
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the year with highest average mpg
best_year = avg_mpg_by_year.idxmax()
print(best_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Drop rows with missing values in relevant columns
df.dropna(subset=['horsepower', 'weight', 'mpg'], inplace=True)

# Get median mpg
median_mpg = df['mpg'].median()

# Filter cars with mpg above median
above_median_mpg = df[df['mpg'] > median_mpg]

# Get the car(s) with max hp_per_weight
max_ratio = above_median_mpg['hp_per_weight'].max()
best_cars = above_median_mpg[above_median_mpg['hp_per_weight'] == max_ratio]

# Print the car name(s)
print(best_cars['name'].values)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')

# Drop missing values
df.dropna(subset=['mpg', 'model_year', 'origin'], inplace=True)

# Group and compute mean
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name, number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# 1. Identify car names appearing in more than one model_year
name_year_counts = df.groupby(['name'])['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index

# 2. Filter df to only those names
df_multi = df[df['name'].isin(multi_year_names)]

# 3. Compute std deviation and mean mpg for each name
stats = df_multi.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean'),
    appearances=('mpg', 'count')
).reset_index()

# 4. Filter to those with std(mpg) < 1.0 and at least 2 appearances
consistent_cars = stats[(stats['mpg_std'] < 1.0) & (stats['appearances'] >= 2)]

# 5. Find the car(s) with the lowest variation in mpg
min_std = consistent_cars['mpg_std'].min()
lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

#Sort by number of appearances (descending), then average mpg (descending)
result = lowest_variation_cars.sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])

# Display relevant columns: model name(s), number of appearances, and average mpg
print(result[['name', 'appearances', 'mpg_mean']])
```
